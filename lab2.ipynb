```python
import pandas as pd

# Creating a sample dataframe with missing values
data = {'A': [1, None, 3, None, 5],
        'B': [None, 2, None, 4, None]}
df = pd.DataFrame(data)

# Applying forward fill to fill missing values
df_ffill = df.ffill()

print("Original DataFrame:")
print(df)
print("\nDataFrame after forward fill:")
print(df_ffill)
```

# 5.backward filling missing values

```python
import pandas as pd

# Creating a sample dataframe with missing values
data = {'A': [1, None, 3, None, 5],
        'B': [None, 2, None, 4, None]}
df = pd.DataFrame(data)

# Applying backward fill to fill missing values
df_bfill = df.bfill()

print("Original DataFrame:")
print(df)
print("\nDataFrame after backward fill:")
print(df_bfill)
```
