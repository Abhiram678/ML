```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
df = pd.read_csv(url, header=None, names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species'])
df
```

# scatter plot


```python
#hue:By specifying the hue parameter, seaborn will automatically assign different colors to the points based on the 
#unique values in the specified column. This makes it easier to visually distinguish between different categories or groups 
#in your data.
sns.scatterplot(x='sepal_length',y='sepal_width',data=df,hue='species')
```

# pair plot

```python
sns.pairplot(df,hue='species')
```

# histogram

```python
df.hist()
```

# kernel density estimation plot

```python
#KDE plots are particularly useful for visualizing the distribution of a single variable. Unlike histograms, which bin the data 
#into intervals and represent the frequency of observations in each interval, KDE plots provide a smooth, continuous 
#representation of the distribution. This can be especially helpful when exploring the shape and characteristics of the 
#distribution, as well as identifying any patterns such as multimodality or skewness.

sns.kdeplot(data=df)
```

# violin plot

```python
sns.violinplot(x='species',y='petal_length',data=df)
```

# joint plot


```python
sns.jointplot(x='petal_length', y='petal_width', data=df, kind='reg')
```

# radviz plot

```python
#Radviz plot is a method of multivariate data visualization that projects data points onto a 2D plane, arranging them around a 
#circle according to the values of their features and connecting them to a central point representing the mean or centroid of 
#the data.
from pandas.plotting import radviz
radviz(df, 'species')
```

# box plot

```python
sns.boxplot(x='species', y='sepal_length', data=df)
```

# contour plot

```python
#A contour plot is a graphical representation of three-dimensional data where levels of a dependent variable are represented by 
#contours on a two-dimensional plane.
sns.kdeplot(x='sepal_length', y='sepal_width', data=df, fill=True)
```

# heat map

```python
a = df.drop('species', axis=1)
sns.heatmap(a.corr(), annot=True, cmap='coolwarm')
```
