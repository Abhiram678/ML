# using z score to identify outliers

```python
#Identifying outliers using z-scores involves calculating the deviation of each data point from the mean in terms
#of standard deviations and marking those beyond a certain threshold as outliers.
from scipy.stats import zscore
z_scores = zscore(df['density (km²)'])
outliers = (z_scores > 3) | (z_scores < -3)
df = df[~outliers]
df
```

```python
import pandas as pd
from scipy.stats import zscore

# Sample DataFrame
data = {'Region': ['Region A', 'Region B', 'Region C', 'Region D', 'Region E'],
        'Density (km²)': [1000, 1200, 1500, 2000, 25000]}  # Note the outlier value (25000)
df = pd.DataFrame(data)

# Identifying outliers using z-scores
z_scores = zscore(df['Density (km²)'])
outliers = (z_scores > 3) | (z_scores < -3)

# Removing outliers from the DataFrame
df = df[~outliers]

# Displaying the updated DataFrame
print(df)
```
